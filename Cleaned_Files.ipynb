```python
# Import the dependencies 
import pandas as pd
```

```python
# Read in the daily_obs file
file = "Resources2/daily_obs.csv"
daily_obs = pd.read_csv(file)
daily_obs.head()
```

```python
# Create new DataFrame with only important columns from the daily_obs DataFrame
daily_obs_clean = daily_obs[['obs_date_time', 'obs_location', 'sky_cover', 'precip_type',
                            'air_temp_min', 'air_temp_max', 'air_temp_current',
                            'snow_height', 'new_snow_height', 'wind_direction', 'wind_speed',
                            'wind_gust', 'hazard']]
```

```python
# Show DataFrame
daily_obs_clean.head()
```

```python
# Drop Null value
daily_obs_clean = daily_obs_clean.dropna()
```

```python
# Check to make sure Null values were dropped
daily_obs_clean.isnull().sum()
```

```python
# Save new DataFrame as CSV file
daily_obs_clean.to_csv('Cleaned_Files/daily_obs_cleaned.csv', index=False)
```
